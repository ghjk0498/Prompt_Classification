```python
ls = ['대분류', '중분류', '소분류', '세분류']
remove_duplicate_set = set()

with open('gpt4-answer.txt', 'r') as txtfile:
    before_level = -1
    ls_html = ''
    b_last = False
    for row in txtfile.readlines():
        level = ls.index(row[:3])
        i = row.find(':') + 2
        title = row[i:]

        if title in remove_duplicate_set:
            continue

        # <ul>
        if before_level < level and not b_last:
            ls_html += ('\t' * level) + '<ul>\n'
            if level == len(ls) - 1:
                b_last = True
        # </ul>
        elif before_level > level:
            for i in range(before_level, level, -1):
                ls_html += ('\t' * i) + '</ul>\n'
            if b_last:
                b_last = False
        
        ls_html += ('\t' * (level + 1)) + '<li>' + title.rstrip() + "</li>\n"

        before_level = level
        remove_duplicate_set.add(title)
    
    # last </ul>
    for i in range(before_level, -1, -1):
        ls_html += ('\t' * i) + '</ul>\n'
    if b_last:
        b_last = False
    
    print(ls_html)
    # print(len(remove_duplicate_set))
```
